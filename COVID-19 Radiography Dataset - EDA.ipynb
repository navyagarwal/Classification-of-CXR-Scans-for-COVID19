```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
covid_path = "Covid-19_Radiography_Dataset\\COVID\\images\\"
viral_path = "Covid-19_Radiography_Dataset\\Viral Pneumonia\\images\\"
normal_path = "Covid-19_Radiography_Dataset\\Normal\\images\\"

num_covid = len(os.listdir(covid_path))
num_viral = len(os.listdir(viral_path))
num_normal = len(os.listdir(normal_path))
```

```python
plt.figure(figsize=(6, 6))   #creates a figure area of 6x6 inch
for i in range(1, 10):
    plt.subplot(3, 3, i)    #divides the area into 3 rows and 3 columns, indexing in subplots starts from 1
    plt.xticks([])  #removes indexing on x axis
    plt.yticks([])  #removes indexing on y axis
    if(i <= 3):
        img = cv2.imread(normal_path + "Normal-" + str(i) + ".png")
        plt.imshow(img)
        plt.xlabel("Normal")
    if(3 < i <= 6):
        img = cv2.imread(covid_path + "COVID-" + str(i) + ".png")
        plt.imshow(img)
        plt.xlabel("Covid")
    if(6 < i <= 9):
        img = cv2.imread(viral_path + "Viral Pneumonia-" + str(i) + ".png")
        plt.imshow(img)
        plt.xlabel("Viral Pneumonia")
plt.show()
```
